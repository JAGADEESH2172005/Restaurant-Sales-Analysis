```python
df['createdTime'] = pd.to_datetime(df['createdTime'])
df['lastUpdatedTime'] = pd.to_datetime(df['lastUpdatedTime'])
```

```python
def time_slot(hour):
    if 0 <= hour < 6:
        return 'Late Night'
    elif 6 <= hour < 12:
        return 'Morning'
    elif 12 <= hour < 17:
        return 'Afternoon'
    elif 17 <= hour < 22:
        return 'Evening'
    else:
        return 'Night'

df['hour'] = pd.to_datetime(df['createdTime']).dt.hour
df['time_slot'] = df['hour'].apply(time_slot)
```

```python
df['hour'] = pd.to_datetime(df['createdTime']).dt.hour
df['date'] = pd.to_datetime(df['createdTime']).dt.date
df['day_name'] = pd.to_datetime(df['createdTime']).dt.day_name()
df['month'] = pd.to_datetime(df['createdTime']).dt.month_name()
df[['createdTime','date','hour','day_name','month','time_slot']].head()
```
